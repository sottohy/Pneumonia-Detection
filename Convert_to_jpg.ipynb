```python
!pip install pydicom

import os
import shutil
import sys
import numpy as np
import pandas as pd
import pydicom
import cv2
```

```python
dataset_path = '/content/drive/MyDrive/study/dataset'

labels = pd.read_csv(os.path.join(dataset_path, "train_labels.csv"))
labels.head()
```

```python
# Convert from DICOM to JPG

train_path = '/content/drive/MyDrive/study/dataset/train_images'
output_train_path = '/content/drive/MyDrive/study/out'

def convert(row, folder_path, output_path):
    # Path to the DICOM file
    file = os.path.join(folder_path, f"{row['patientId']}.dcm")
    if not os.path.isfile(file):
        print(f"File not found: {file}")
        return

    # Read the DICOM file
    image = pydicom.dcmread(file)

    # Get the pixel array from the DICOM file
    image = image.pixel_array

    image = cv2.resize(image, (224,224))

    # Convert the image to RGB
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

    # The output path for the JPG file
    out_path = os.path.join(output_path, f"{row['patientId']}.jpg")

    # Write the image to the output path as a JPG file
    cv2.imwrite(out_path, image)

for _ , row in labels.iterrows():
    convert(row, train_path, output_train_path)
```
